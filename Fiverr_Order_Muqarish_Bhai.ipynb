```python
from selenium import webdriver
import time
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
#driver.get("https://int.soccerway.com/")
```

```python
click_icon=driver.find_elements(By.CLASS_NAME,'livescores-comp')
for i in click_icon:
    button = i.find_element(By.CLASS_NAME,"expand-icon")
    button.click()
```

```python
teams_links=[]
teams=driver.find_elements(By.CLASS_NAME,'teams')

for i in teams:
    anchor=i.find_elements(By.TAG_NAME,'a')
    for j in anchor:
        href=j.get_attribute('href')
        print(href)
        teams_links.append(href)
```

```python
len(teams_links)
```

```python
#import pandas as pd
```

```python
all_links=pd.DataFrame({
    'Links_Teams':teams_links})

all_links.to_csv('teams_links.csv')
```

```python
driver.get('https://int.soccerway.com/matches/2023/09/06/world/friendlies/macao/bhutan/4213597/')
```

```python
click_1=driver.find_element(By.XPATH,'//*[@id="page_match_1_block_match_team_matches_22"]')
click_1.click()
```

```python
home_click=driver.find_element(By.ID,'page_match_1_block_match_team_matches_22_2_2')
home_click.click()
```

```python
home_date=driver.find_elements(By.XPATH,'/html/body/div[8]/div[2]/div[2]/div/div/div[2]/div[3]/div[6]/div/div/div[3]/div[2]/table')
home_date_full=[]
home_competition=[]
home_team_a=[]
home_team_b=[]

for i in home_date:
    date=i.find_elements(By.CLASS_NAME,'full-date')
    competition=i.find_elements(By.CLASS_NAME,'competition')
    team_a=i.find_elements(By.CLASS_NAME,'team.team-a')
    team_b=i.find_elements(By.CLASS_NAME,'team.team-b')
    score=i.find_elements(By.CLASS_NAME,'score-time.score')

    for j,k,l,m in zip(date,competition,team_a,team_b):
        home_date_full.append(j.text)
        home_competition.append(k.text)
        home_team_a.append(l.text)
        home_team_b.append(m.text)
        
```

```python
home_competition,home_date_full,home_team_a
```

```python
home_score=[]
for i in home_date:
    score=i.find_elements(By.CLASS_NAME,'score-time.score')
    
    for j in score:
        home_score.append(j.text)

print(home_score)
```

```python
home_1=pd.DataFrame({
    'home_date':home_date_full[1:6],'home_competition':home_competition[1:6],'home_team_a':home_team_a[1:6],'home_team_b':home_team_b[1:6],'home_score':home_score
})
home_1.to_csv('home_3.csv')
```
